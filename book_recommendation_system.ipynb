## Collaborative filtering recommender system

- It is a techique in which a product is recommended to the user which was used by some other similar users
- The data should contain set of items and set of users who have reacted to some items
- The reaction can be explicit(ratings on some scale like 1 to 5, likes or dislikes) and implicit((viewing an item, adding it to a wish list, the time spent on an item)
- when we want to give the recommendation to the user we will find the similarity between the users
- Now in order to find the similarity between users , Make a graph where different item are our axes, and the users will be the points in the graph 


```python
df_ratings_name = df_ratings_name.copy()
```

```python
df_ratings_name.head()
```

```python
# finding the user who have voted more than 250 times
x = df_ratings_name.groupby('User-ID').count()['Book-Rating']>250
high_level_user = x[x].index
new_df = df_ratings_name[df_ratings_name['User-ID'].isin(high_level_user)]
```

```python
new_df.shape
```

```python
# finding the books which have more than 50 votes
y = new_df.groupby('Book-Title').count()['Book-Rating'] > 50
high_level_book = y[y].index
new_df = new_df[new_df['Book-Title'].isin(high_level_book)]
```

```python
new_df.shape
```

```python
# creating the pivot table for applying cosine similarity
pt = new_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
# finding the cosine similarity i.e similar books
# since our row is vectors of book we will get the similar score for book i.e (item-based )if the our row was of user we will get 
# similar scores for user i.e user-based
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(pt)
similarity_scores.shape
# similarity_scores[0] 1984
```

```python
# here each array(similarity_scores[0]) tells similarity scores with every books 
# for example similarity_scores[0] tells us how similar is 1984 with all other books
```

```python
def recommend(book_name):
    # fetching the index
    index = np.where(pt.index==book_name)[0][0]
    score = similarity_scores[index]
    index_score = list(enumerate(score))
    top_books = sorted(index_score,key=lambda x:x[1],reverse = True)[1:6]
    
    data = []
    for i in top_books:
        book_info = []
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        # same books might have the different year of publication and different ISBN, but we only need one book
        temp_df = temp_df.drop_duplicates('Book-Title')
        
        book_info.append(temp_df['Book-Title'].values[0])
        book_info.append(temp_df['Book-Author'].values[0])
        book_info.append(temp_df['Year-Of-Publication'].values[0])
        book_info.append(temp_df['Image-URL-M'].values[0])
        
        data.append(book_info)
    
    return data

# data should be the list of lists because in templates we have to use the loop to fetch the details of a book
    
    
    
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(pt_user)
```

```python
def find_similar_user(user_id):
    index = np.where(pt_user.index==user_id)[0][0]
    score = user_similarity[index]
    index_score = list(enumerate(score))
    top_users = sorted(index_score,key=lambda x:x[1],reverse=True)[1:6]
    
    similar_users = []
    for i in top_users:
        similar_users.append(pt_user.index[i[0]])
    
    return similar_users
```

```python
def recommend_books(users):
    books=[]
    for i in users:
        temp_df = df_ratings_name[df_ratings_name['User-ID']==i]
        max_ratings = temp_df['Book-Rating'].max()
        top_preference = temp_df[temp_df['Book-Rating']==max_ratings].iloc[0]
        cols = list(top_preference[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M']].values)
        books.append(cols)
        
    return books
```

## content based filtering

```python
temp_df = df_ratings_name.groupby('Book-Title').count()['Book-Author'].reset_index()
temp_df = temp_df.rename(columns={'Book-Author':'views'})
temp_df.head(2)
```

```python
df_views = df_ratings_name.copy()
```

```python
df_views = df_views.merge(temp_df,on=['Book-Title'])
```

```python
#This dataframe also contains the no. of view of every books
df_views.head()
```

```python
author =df_views[df_views['Book-Title'] == 'Zoya']['Book-Author'].values[0]
```

```python
author_books = df_views[(df_views['Book-Author'] == author) & (df_views['views']>50)]
author_books
```

```python
avg_rating = author_books.groupby('Book-Title').mean(numeric_only=True)
```

```python
# famous_books=list(avg_rating[avg_rating['Book-Rating']>7.index)
famous_books=list(avg_rating[avg_rating['Book-Rating']>2.5].index) 
# the book same as user input might alos be recommended so filtering it
famous_books =[book for book in famous_books if book!='Zoya'] 
famous_books
```

```python
author_books[author_books['Book-Title'].isin(famous_books)][['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','views']].drop_duplicates('Book-Title')
```

```python
def content_recommend(book_name):
    #finding the name of the author
    author =df_views[df_views['Book-Title'] == book_name]['Book-Author'].values[0]
    # finding books written by author and the books have the views more than 50
    author_books = df_views[(df_views['Book-Author'] == author) & (df_views['views']>50)]
    
    #finding the avg-ratings for all books
    avg_rating = author_books.groupby('Book-Title').mean(numeric_only=True)
    
    famous_books=list(avg_rating[avg_rating['Book-Rating']>2.5].index) 
    # the book same as user input might also be recommended so filtering it
    famous_books =[book for book in famous_books if book!=book_name] 
    
    #finally returning the rows which has the famoust books and droping the duplicates since many user have read the same book
    suggestions = author_books[author_books['Book-Title'].isin(famous_books)][['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','views']].drop_duplicates('Book-Title').values
    
    return suggestions
```

```python
content_recommend('The Lovely Bones: A Novel')
```

- I want to recommed the books on the basis of same author.
- If i simply recommed the books that the same author has written than there will be 100's of books. I can't just simply recommend the 100 books to the user
- so i did the filtering on the basis of ratings. I only recommended the books whose ratings are above the certain threshold.
- There is a problem if the books is rated high (9 or 10) only by 1 user due to which average rating will be high then my system will recommend the unpopular books which can be risky to recommend. since many users has not read it.
- so before finding the average rating i need to filter the books on the basis of no.of votes to do this i added the views column in the dataset
- If the author of the book is not very famous than there might not be any books to recommend
